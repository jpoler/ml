```python
%matplotlib inline
import matplotlib.pyplot as plt  # type: ignore
import numpy as np
import numpy.typing as npt

from data.sin import sin_curve, sin_data
from linear_regression import GaussianBasisLeastSquaresRegression, PolynomialBasisLeastSquaresRegression
from grid.hyperparameters import class_partially_bound_parameters, plot_class_vs_parameter, PlotData
```

```python
s_curve = sin_curve()
s_data = sin_data(n_train=200, n_test=100)
```

```python
model = PolynomialBasisLeastSquaresRegression(m_degrees=10)
model.fit(s_data.x_train, s_data.y_train)

y_predictions = model.predict(s_data.x_test)

plt.plot(
    s_curve.x,
    s_curve.y,
    "b-",
    s_data.x_train,
    s_data.y_train,
    "go",
    s_data.x_test,
    y_predictions,
    "ro",
)
```

```python
model = PolynomialBasisLeastSquaresRegression(
    m_degrees=10, regularization_coefficient=3.0
)
model.fit(s_data.x_train, s_data.y_train)

y_predictions = model.predict(s_data.x_test)

plt.plot(
    s_curve.x,
    s_curve.y,
    "b-",
    s_data.x_train,
    s_data.y_train,
    "go",
    s_data.x_test,
    y_predictions,
    "ro",
)
```

```python
model = GaussianBasisLeastSquaresRegression(
    low=0.0, high=2*np.pi, num=10, stddev=.6
)
model.fit(s_data.x_train, s_data.y_train)

y_predictions = model.predict(s_data.x_test)

plt.plot(
    s_curve.x,
    s_curve.y,
    "b-",
    s_data.x_train,
    s_data.y_train,
    "go",
    s_data.x_test,
    y_predictions,
    "ro",
)
```

```python
gaussian_partial = class_partially_bound_parameters(GaussianBasisLeastSquaresRegression, low=0.0, high=2*np.pi, num=10, stddev=.6)
polynomial_partial = class_partially_bound_parameters(PolynomialBasisLeastSquaresRegression, m_degrees=10)
# todo fix this with subclassing
plot_data = PlotData(x_train=s_data.x_train, y_train=s_data.y_train, x_test=s_data.x_test, y_test=s_data.y_test)
plot_class_vs_parameter(plt, [gaussian_partial, polynomial_partial], "regularization_coefficient", [None, 0.5, 1.0, 5.0, 10.0, 100.0], plot_data)
```
