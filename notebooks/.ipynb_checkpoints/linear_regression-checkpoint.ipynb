```python
%matplotlib inline
import matplotlib.pyplot as plt  # type: ignore
import numpy as np
import numpy.typing as npt

from data.sin import sin_curve, sin_data
from linear_regression import GaussianBasisLeastSquaresRegression, PolynomialBasisLeastSquaresRegression
from grid.hyperparameters import full_data_grid, ParameterSpace, plot_predictions
```

```python
s_curve = sin_curve()
s_data = sin_data(n_train=200, n_test=100)
```

```python
model = PolynomialBasisLeastSquaresRegression(m_degrees=10)
model.fit(s_data.x_train, s_data.y_train)

y_predictions = model.predict(s_data.x_test)

plt.plot(
    s_curve.x,
    s_curve.y,
    "b-",
    s_data.x_train,
    s_data.y_train,
    "go",
    s_data.x_test,
    y_predictions,
    "ro",
)
```

```python
model = PolynomialBasisLeastSquaresRegression(
    m_degrees=10, regularization_coefficient=3.0
)
model.fit(s_data.x_train, s_data.y_train)

y_predictions = model.predict(s_data.x_test)

plt.plot(
    s_curve.x,
    s_curve.y,
    "b-",
    s_data.x_train,
    s_data.y_train,
    "go",
    s_data.x_test,
    y_predictions,
    "ro",
)
```

```python
model = GaussianBasisLeastSquaresRegression(
    low=0.0, high=2*np.pi, num=10, stddev=.6
)
model.fit(s_data.x_train, s_data.y_train)

y_predictions = model.predict(s_data.x_test)

plt.plot(
    s_curve.x,
    s_curve.y,
    "b-",
    s_data.x_train,
    s_data.y_train,
    "go",
    s_data.x_test,
    y_predictions,
    "ro",
)
```

```python
models = [PolynomialBasisLeastSquaresRegression,PolynomialBasisLeastSquaresRegression]
param_spaces = [
    ParameterSpace(keyword="regularization_coefficient", space=np.linspace(0.0, 1000.0, 10)),
    ParameterSpace(keyword="m_degrees", space=[i for i in range(10)])
]

grid = full_data_grid(models, s_data, {"regularization_coefficient": 1., "m_degrees": 5}, param_spaces)

plot_predictions(plt, list(grid))
```
